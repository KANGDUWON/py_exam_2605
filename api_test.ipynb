```python
import requests as req

url = 'https://jsonplaceholder.typicode.com/posts/1'

res = req.get(url)


print("== 상태코드 ==")
print(res.status_code)

print("== json 형태 ==")
print(res.json())

print("== 실제 데이터 ")
print(res.text)


data = res.json()
print('제목:',data['title'])
print('내용:',data['body'])

data2 = {
    "title": "test",
    "body": "hello"
}
res2 = req.post(url, json=data)
print(res2.json())
```
